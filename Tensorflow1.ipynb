Train Artificial Neural Network odel using Tensor flow using ionoshere dataset


```python
import pandas as pd
url='https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/Ionosphere.csv'
df=pd.read_csv(url)
df.head()
```

```python
x=df.drop(['Class'],axis=1)
y=df['Class']
print(y)
```

```python
from os import XATTR_CREATE
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

Hidden layers(neuron modle)

```python
from keras.models import Sequential,load_model
from keras.layers import Dense,Dropout
#import dependancies
```

```python
#build model

model=Sequential()
model.add(Dense(units=10,activation='relu',input_dim=len(x_train.columns)))
#relu normal needs in hidden layers
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.summary()
```

```python
#compile the model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
#train method
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,batch_size=32)
```

```python
model.save('ionosphere.h5')
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

**#create new filtered data set using columns('N_total', 'N_bulk', 'N_surface', 'R_avg')**

```python
filter_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

**# read sample with first 20 data from filtered data set**

```python
filter_df.head(20)
```

```python
cols = filter_df.columns

def find_mean(cols):
  print("================== mean values===============\n")
  for x in cols:
      print(x + 'mean = ',df[x].mean())

def find_sd(cols):
  print("\n================== SD values=================\n")
  for x in cols:
      print( x + 'sd = ',df[x].std())


find_mean(cols)
find_sd(cols)
```

```python
cols = filter_df.columns

def find_quartiles(cols):
  print("================== quartile values===============\n")
  for x in cols:
      print("quartile values of " + "'" + x + "'\n")
      print('q1 = ',df[x].quantile(0.25))
      print('q2 = ',df[x].quantile(0.5))
      print('q3 = ',df[x].quantile(0.75))
      print("\n")

find_quartiles(cols)
```

```python
cols = filter_df.columns

def display_hist(cols):
  for x in cols:
      plt.figure(figsize=(5,4))
      plt.hist(df[x])
      plt.title("Histogram of " + "'" + x + "'" + " fetaure")
      plt.xlabel(x)
      plt.ylabel("Frequency")
      plt.show()
      print("\n")

display_hist(cols)
```

```python
cols = filter_df.columns

def display_hist(cols):
  for x in cols:
      plt.figure(figsize=(5,4))
      sns.histplot(df[x])
      plt.title("Histogram of " + "'" + x + "'" + " fetaure")
      plt.xlabel(x)
      plt.ylabel("Frequency")
      plt.show()
      print("\n")

display_hist(cols)
```

```python
plt.figure(figsize=(20,30))
sns.pairplot(data=filter_df)
```
